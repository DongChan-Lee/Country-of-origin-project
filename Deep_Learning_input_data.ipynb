# 3. X값 만들기

### 2차원 X값 먼저 만들기

```python
%%time
total_input = []   
for i in range(16235):  # 고유한 grid 개수
    print('Process : {} | Total : 16235'.format(i+1), end='\r')
    master_table_clip = globals()['master_table_{}'.format(i+1)]   # master_table 1개씩 master_table_clip 변수에 저장
    clip = data2[(data2['id']==unique_input['id'].iloc[i])&(data2['행정동']==unique_input['행정동'].iloc[i])]   # 고유한 grid id, 행정동에 맞는 데이터들을 clip 변수에 저장
    for j in range(len(clip)):  # clip의 행의 개수
        mm = length_adjust(str(clip.iloc[j,2]))  # 조사월 형태 맞추기(ex. 6 → 06, 7 → 07)
        ym = str(clip.iloc[j,1]) + mm    # (ex. 2010 + 01 → 201001)
        master_table_clip.loc[ym, :] = data2_scaled[clip.index[j]]    # scaling한 X_column의 값
    total_input.append(master_table_clip.values[:108, :])
```

# 4. Y값 만들기
### 2019년 적발건수

```python
data2 = pd.read_csv('최종_데이터프레임_데이터정제.csv', encoding='cp949', index_col=0)
data2
```

```python
# Y(적발건수) 만드는 버전
data2 = data2[['id', '조사년도', '조사월', '행정동',
               '총적발건수']]
data2
```

```python
data2['총적발건수'].unique()
```

```python
def length_adjust(str_) :                   # if month, day is single number, convert to two-numbers (ex. 6 → 06, 7 → 07)
    return (2-len(str_)) * '0' + str_ 

def date_convertor(date) :                  # ex. '2010-01-01' → '20100101'
    year = length_adjust(str(date.year))
    month = length_adjust(str(date.month))
    yyyymm = year+month
    return yyyymm                       # return 'str' type

year_month = pd.date_range(start='20100101', end='20191231', freq='M')  # 20100101 ~ 20191231의 날짜를 월 단위로 생성(201001, 201002, ..., 201912)
year_month = list(map(date_convertor, year_month))  # len(year_month) : 120
temp_columns = data2.columns[4:]  # 1
temp_columns
```

#### 각각의 고유한 grid마다 'master_tabel_1' ~ 'master_table_16235'라는 변수를 생성하여 index는 20100101 ~ 20191231의 날짜, column은 변수인 데이터프레임을 지정해준다.

```python
for i in range(16235):  # 고유한 grid 개수
    globals()['master_table_{}'.format(i+1)] = pd.DataFrame(np.full(shape = (len(year_month), len(temp_columns)), fill_value=-1), index = year_month, columns = temp_columns)
```

```python
master_table_1
```

```python
%%time
# scaling하지 않고 데이터프레임에서 값 가져오기
for i in range(16235):  # 고유한 grid 개수
    print('Process : {} | Total : 16235'.format(i+1), end='\r')
    master_table_clip = globals()['master_table_{}'.format(i+1)]   # master_table 1개씩 master_table_clip 변수에 저장
    clip = data2[(data2['id']==unique_input['id'].iloc[i])&(data2['행정동']==unique_input['행정동'].iloc[i])]   # 고유한 grid id, 행정동에 맞는 데이터들을 clip 변수에 저장
    for j in range(len(clip)):  # clip의 행의 개수
        mm = length_adjust(str(clip.iloc[j,2]))  # 조사월 형태 맞추기(ex. 6 → 06, 7 → 07)
        ym = str(clip.iloc[j,1]) + mm      # (ex. 2010 + 01 → 201001)
        master_table_clip.loc[ym, :] = clip.values[j][4:]  # 총적발건수 값을 추가
```

```python
%%time
y_grid = []  # 완성본 

for i in range(16235):  # 16235 grid별
    sum_ = 0  # 총 적발건수
    print('Process : {} | Total : 16235'.format(i+1), end='\r')
    master_table_clip = globals()['master_table_{}'.format(i+1)]
    
    # 총적발건수 value counts
    unique, counts = np.unique(master_table_clip.iloc[-12:, 0].values, return_counts=True)    # 2019년 총적발건수
    cnt_dict = dict(zip(unique, counts))   # 2019년 총적발건수 {value : count}
    del cnt_dict[-1.0]  # padding값 -1 제거
    
    # 총 적발건수 count
    if len(cnt_dict) == 0:         # padding값 -1밖에 없었다면(모두 결측치라면)
        # 각 grid의 2010~2018년 총적발건수 다 더해서, "조사나간 연도 수"로 나누기
        count = 0  # 조사나간 연도 수
        for k in range(9):
            master_table_year = master_table_clip.iloc[12*k:12*(k+1), :]  # 1년치만 잘라서
            jucval_list = list(master_table_year['총적발건수'].values)
            if all(jucval_list) != -1:  # 전체가 padding값 -1이 아니라면,
                count += 1              # 조사나간 연도수 count + 1
                for char in jucval_list:  # 값을 하나씩 가져와서,
                    if char >= 0:         # padding값이 아니라면,
                        sum_ += char  # 총 적발건수에 더해주기
        y_grid.append(sum_/count)
    else:                          # padding값 -1 이외의 값이 있었다면
        # 각 grid의 2019년 총적발건수 세기
        list_ = []
        for a in cnt_dict:
            list_.append(a * cnt_dict[a])      # a : 적발건수 unique값 --> unique값에, 해당하는 list_2값 곱해서 다 더해주면 "총적발건수"
                                               # 즉, a * cnt_dict1[a] 한 것들을 다 더해주면 "총적발건수"
        sum_ += np.sum(list_)
        y_grid.append(sum_)
```

```python
y_grid = np.array(y_grid)
y_grid = y_grid.reshape(16235, 1)
y_grid.shape
```

```python
y_grid_df = pd.DataFrame(y_grid, columns=['총적발건수'])
y_grid_df
```

```python
final_scaler_y = MinMaxScaler(feature_range=(0,1))  # normalize the label
final_scaler_y.fit(y_grid_df)
y_scaled = final_scaler_y.transform(y_grid_df)
y_scaled
```
